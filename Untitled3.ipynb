```python
import pandas_ta as ta
import yfinance as yf
```

```python
data = yf.download(tickers='AAPL',start = '2021-01-01', end = '2023-01-1')
```

```python
data.columns = ['_'.join(map(str, col)) for col in data.columns.to_flat_index()]
```

```python
data
```

```python
data.columns = ['Close','High','Low','Open','Volume']
```

```python
data['ema_5'] = ta.ema(data['Close'],length=5, append = True, adjust = True)
```

```python
data
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('dark_background')
```

```python
plt.figure(figsize=(10,6))
plt.plot(data['Close'],label = ['Close Price'],linewidth = 2)
plt.plot(data['ema_5'],label = ['5-day EMA'],linestyle = '--',linewidth = 2)
plt.title('AAPL Price with 5 day EMA')
plt.grid()
plt.legend()
plt.show()
```

```python
data['sma_50'] = ta.sma(data['Close'],lenght = 50)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data['Close'],label = ['Close Price'],linewidth = 2)
plt.plot(data['sma_50'],label = ['50-day SMA'],linestyle = '--',linewidth = 2)
plt.title('AAPL Price with 50 day SMA')
plt.grid()
plt.legend()
plt.show()
```

```python
data['chaikin'] = ta.adosc(high=data['High'], low = data['Low'],volume=data['Volume'],close = data['Close'],fast = 3, slow = 10)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data['Close'],label = ['Close Price'],linewidth = 2)
plt.plot(data['chaikin'],label = ['chaikin'],linestyle = '--',linewidth = 2)
plt.title('chaikin')
plt.grid()
plt.legend()
plt.show()
```
