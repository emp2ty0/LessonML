```python
def education_group(x):
  if x in ['High School','Primary School']:
    return 'School'
  if x in ['Masters','Phd']:
    return 'Masters_and_Phd'
  return x
```

```python
df['education_group'] = df['education_level'].apply(education_group)
```

```python
df
```

```python
df.groupby('education_group',dropna=False)['target'].agg(['count','mean'])
```

```python
def new_training_hours(row):
  if row['education_level'] == 'Phd':
    return row['training_hours'] + 1000
  return row['training_hours']
```

```python
df['new_training_hours'] = df.apply(new_training_hours,axis=1)
```

```python
def experience_int(x):
  if x == '>20':
    return 999
  if x == '<1':
    return 0
  return int(x)
```

```python
df['experience'] = df['experience'].fillna(-1)
```

```python
df['experience_int'] = df['experience'].apply(experience_int)
```

```python
df
```

```python
df.groupby('experience_int')['target'].agg(['count','mean']).sort_values('mean').head()
```

```python
df.plot(x = 'city_development_index', y = 'experience_int',kind = 'scatter')
```
